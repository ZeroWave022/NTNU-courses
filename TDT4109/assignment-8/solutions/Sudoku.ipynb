I denne oppgaven skal vi først lage et sudoku-spill (før vi lager en sudoku-løser om noen år). Om du ikke kjenner reglene til sudoku kan du lese deg opp på de selv [her](https://no.wikipedia.org/wiki/Sudoku). Del 1 tar seg av å lage et spillbart sudoku-brett.

Du står fritt til å bygge opp brettet slik som du vil, men brettet skal oppfylle følgende krav:

* Brukeren skal kunne skrive inn et tall i en valgfri celle. Dersom tallet ikke er gyldig, dvs. ikke mellom 1 og 9, skal en feilmelding skrives ut.
* Brukeren skal ikke kunne fylle inn et tall som allerede finnes i den samme raden, kolonnen eller kvadratet.
* Brukeren skal kunne slette et tall fra en celle.
* Hver gang brukeren fyller inn eller sletter et tall skal det nye brettet skrives ut på en fin måte. Et eksempel kan være som vist under (tallene over og ved siden av brettet angir her henholdsvis kolonne- og radnummer).
* Brukeren skal kunne laste inn et brett fra en tekstfil.
* Et halvutfylt brett skal kunne lagres til fil, slik at man kan fullføre det senere.
* Spillet skal skrive ut en hyggelig gratulasjonsmelding dersom man har klart brettet.
* Alt skal utføres gjennom et brukervennlig grensesnitt. Det vil si at brukeren ikke skal trenge å kalle på funksjonene selv, men at alt gjøres via input eller ved å lese fra/skrive til fil.

Eksempel på brett:
```python
    0 1 2   3 4 5   6 7 8 
  +-------+-------+-------+
0 | 0 0 6 | 9 0 5 | 0 1 0 |
1 | 9 7 0 | 0 1 2 | 3 0 5 |
2 | 0 2 0 | 0 0 4 | 8 6 0 |
  +-------+-------+-------+
3 | 5 0 3 | 8 0 0 | 0 2 0 |
4 | 0 0 0 | 0 0 0 | 0 0 0 |
5 | 0 8 0 | 0 0 1 | 9 0 7 |
  +-------+-------+-------+
6 | 0 5 4 | 1 0 0 | 0 7 0 |
7 | 2 0 7 | 4 5 0 | 0 9 3 |
8 | 0 6 0 | 7 0 3 | 1 0 0 |
  +-------+-------+-------+
```

````python
from random import randint

def get_box(board, row, col):
    if 0 <= row <= 2:
        box_rows = (0, 3)
    elif 3 <= row <= 5:
        box_rows = (3, 6)
    elif 6 <= row <= 8:
        box_rows = (6, 9)
    else:
        raise ValueError("Couldn't find box rows for specified row!")
    if 0 <= col <= 2:
        box_cols = (0, 3)
    elif 3 <= col <= 5:
        box_cols = (3, 6)
    elif 6 <= col <= 8:
        box_cols = (6, 9)
    else:
        raise ValueError("Couldn't find box columns for specified column!")
    box = []
    for row in range(box_rows[0], box_rows[1]):
        box.extend(board[row][box_cols[0] : box_cols[1]])
    return box

def number_valid(board, row, col, num):
    """Check if `num` is corrent at `row`, `col` in `board` (assuming 9x9)."""
    row_values = board[row]
    col_values = [r[col] for r in board]
    box = get_box(board, row, col)
    return num not in row_values and num not in col_values and num not in box

def select_board():
    rand = randint(1, 100)
    with open("./sudoku_boards.txt", encoding="utf-8") as f:
        content = f.readlines()
    lines = content[rand * 10 - 10 : rand * 10 - 1]
    lines = [line.replace("\n", "") for line in lines]
    board = []
    for line in lines:
        board.append([int(num) if num != "." else None for num in line])
    return board

def load_board(filename):
    with open(filename, "r", encoding="utf-8") as f:
        content = f.readlines()
    lines = [line.replace("\n", "") for line in content]
    board = []
    for line in lines:
        board.append([int(num) if num != "." else None for num in line])
    return board

def save_board(board, filename):
    content = ""
    for row in board:
        content += "".join(str(num) if num is not None else "." for num in row) + "\n"
    with open(filename, "w", encoding="utf-8") as f:
        f.write(content)

def show_board(board):
    separator = "  +-------+-------+-------+"
    print("    1 2 3   4 5 6   7 8 9")
    print(separator)
    for i, row in enumerate(board, 1):
        row = [str(i) if i is not None else "0" for i in row]
        print(f"{i} | {' '.join(row[0:3])} | {' '.join(row[3:6])} | {' '.join(row[6:9])} |")
        if i % 3 == 0:
            print(separator)

example_board = [
    [6, None, None, None, None, 9, None, None, 2],
    [5, None, None, None, 7, 2, None, 6, 1],
    [2, 1, None, 5, None, None, None, None, None],
    [7, None, None, 2, None, None, 5, 4, 9],
    [9, 3, 5, None, 4, None, None, 1, 8],
    [1, None, 2, 9, None, 8, 6, None, None],
    [None, None, 9, 1, 2, None, None, None, 5],
    [3, None, None, None, None, 4, None, None, None],
    [8, None, None, 7, None, 5, 3, 9, 4],
]

def main():
    print("Velkommen!")
    load = input("Ønsker å du å laste inn et tidligere spill? (ja/nei): ").lower() == "ja"
    if load:
        board = load_board("./sudoku_in_progress.txt")
        print("Brett lastet inn!")
    else:
        board = select_board()
    # board = example_board
    print('Husk at du kan når som helst skrive "lagre" for å lagre spillet til en tekstfil, og fortsatte senere!')

    while True:
        show_board(board)
        delete = input('Dersom du vil fjerne en celle, skriv "fjern" (ellers trenger du ikke skrive noe som helst): ').lower()
        row = input("Skriv inn raden der du vil plassere/fjerne et tall (1-9): ")
        col = input("Skriv inn kolonnen der du vil plassere/fjerne et tall (1-9): ")

        if any(i == "lagre" for i in [delete, row, col]):
            save_board(board, "./sudoku_in_progress.txt")
            print("Sudoku lagret!")
            break

        row = int(row) - 1
        col = int(col) - 1

        if board[row][col] is not None:
            if delete == "fjern":
                print(f"Tallet {board[row][col]} har blitt fjernet.")
                board[row][col] = None
                continue
            print(f"Tallet {board[row][col]} finnes på denne plasseringen allerede!")
            continue

        num = int(input(f"Skriv inn tallet du vil plassere på ({row + 1}, {col + 1}): "))
        if row not in range(0, 9) or col not in range(0, 9):
            print("Feil rad eller kolonne!")
            continue
        if num not in range(1, 10):
            print("Tallet må være mellom 1 og 9!")
            continue
        if number_valid(board, row, col, num):
            print("Riktig! Tallet ble plassert.")
            board[row][col] = num
        else:
            print("Feil! Prøv igjen.")
        
        if all(None not in row for row in board):
            print("Gratulerer, du har fullført brettet!")
            break

if __name__ == "__main__":
    main()
````
